## MatPlotLib

```python
import numpy as np
import matplotlib.pyplot as plt

x = range(1000)             # hodnoty na ose x
y = [i ** 2 for i in x]     # hodnoty na ose y

plt.plot(x,y)               # vykreslení průběhu funkce
plt.show()                  # zobrazení grafu
```

```python
x = np.linspace(0,100)
y1 = 100 * x ** 2           # hodnoty 1. funkce
y2 = x ** 3                 # hodnoty 2. funkce

plt.plot(x, y1)             # vykreslení průběhu 1. funkce
plt.plot(x, y2)             # vykreslení průběhu 2. funkce
plt.show();
```

```python
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

plt.plot(x, y)              # vykreslení

plt.xlabel("x")             # popis osy x
plt.ylabel("sin(x)")        # popis osy y
plt.grid()                  # mřížka

plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, x, y2)          # vykreslení průběhu obou funkcí
 
plt.xlabel("x")                 # popis osy x
plt.ylabel("sin(x) a cos(x)")   # popis osy y
 
plt.show()
```

```python
x = np.linspace(0.01, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x)/x
 
# vykreslení funkcí se změnou stylu
plt.plot(x, y2, "r.", label="cos")      # červená tečkovaná
plt.plot(x, y3, "g--", label="sinc")    # zelená čárkovaná
plt.plot(x, y1, "b-", label="sin")      # modrá plná

plt.legend(loc="lower left")            # přidání legendy

plt.xlabel("x")                         # popis osy x
plt.ylabel("sin(x), cos(x) a sinc(x)")  # popis osy y
 
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.sin(3*x)/(x+1)

plt.fill(x, y1, "red", x, y2, "yellow", alpha=0.3)  # alfa kanál (1 neprůhledné)
#plt.plot(x, y1, "red", x, y2, "yellow")             # obtáhne obrysy
 
plt.xlabel("x")                                     # popis osy x
plt.ylabel("sin(x) a sinc(3x)")                     # popis osy y
 
plt.show()
```

```python
plt.plot(x, y1, "b-", label="sin")
plt.plot(x, y2, "g--", label="sinc")

plt.fill(x, y1, "orange", x, y2, "blue", alpha=0.2)     # alfa kanál (1 neprůhledné)

plt.legend(loc="lower left")                            # přidání legendy

plt.show()
```

```python
x = np.linspace(0.001, 2*np.pi, 100)
y1 = np.sin(5*x)
y2 = np.sin(5*x)/(x+1/2)                        # sinus cardinalis
y3 = 1/(x+1/2)
y4 = -y3
 
plt.figure(figsize=(10, 8), dpi=144)            # velikost obrázku (")

plt.fill(x, y1, "yellow", alpha=0.3, label="sin x")
plt.fill(x, y2, "r.", alpha=0.6, label="sinc 5x")
plt.plot(x, y3, "g--", label="obalka sinc")
plt.plot(x, y4, "g--", label="obalka sinc")

plt.legend(loc="upper right")                   # přidání legendy

plt.xlabel("x")                                 # popis osy x
plt.ylabel("sin(x) a sinc(3x)")                 # popis osy y
 
plt.show()
```

Zobrazení os, mřížky, nastavení rozsahu, popisky

```python
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, "b-", label="sin")
plt.plot(x, y2, "r-", label="cos")

plt.legend(loc="lower left")                    # přidání legendy

plt.axis([-1, 8, -1.5, 1.5])                    # nastavení rozsahů (xmin, xmax, ymin, ymax)

plt.grid(True)                                  # povolení zobrazení mřížky
plt.axhline(y = 0, color='k')                   # vykreslení osy x (černě)
plt.axvline(x = 0, color='k')                   # vykreslení osy y (černě)

plt.xlabel("x")                                 # popis osy x
plt.ylabel("sin(x) a cos(x)")                   # popis osy y

# vložit první popisek do grafu
plt.annotate("maximální hodnota sin(x)",
             xy=(np.pi/2, 1.0),                 # poloha šipky (x, y)
             xytext=(1, 1.3),                   # poloha textu (x, y)
             arrowprops=dict(arrowstyle="->"))  # typ šipek
 
# vložit druhý popisek do grafu
plt.annotate("minimální hodnota cos(x)",
             xy=(np.pi, -1.0),                  # poloha šipky (x, y)
             xytext=(2, -1.3),                  # poloha textu (x, y)
             arrowprops=dict(arrowstyle="<-"))  # typ šipek
 
plt.show()
```

```python
x = [1, 2, 3, 4]
plt.plot(x, [1, 4, 9, 16], 'ro')                # red, kroužky
plt.plot([3, 4], [8, 10], 'b+',  markersize=8)  # blue, křížky, velikost
plt.plot(x, [4, 5, 6, 7], 'm:')                 # magenta, tečky
plt.ylabel('čísla')
plt.show()
```

```python
# b     blue          .     point              -     solid
# g     green         o     circle             :     dotted
# r     red           x     x-mark             -.    dashdot 
# c     cyan          +     plus               --    dashed   
# m     magenta       *     star             (none)  no line
# y     yellow        s     square
# k     black         d     diamond
# w     white         v     triangle (down)
#                     ^     triangle (up)
#                     <     triangle (left)
#                     >     triangle (right)
#                     p     pentagram
#                     h     hexagram
```

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(9, 3))      # velikost obrázku

plt.subplot(131)                # rozložení 1 × 3, 1. graf
plt.bar(names, values)          # sloupcový

plt.subplot(132)                # rozložení 1 × 3, 2. graf
plt.scatter(names, values)      # bodový

plt.subplot(133)                # rozložení 1 × 3, 3. graf
plt.plot(names, values)         # spojnicový

plt.suptitle('Categorical Plotting')
plt.show()
```

## Způsoby přístupu

```python
x = np.linspace(0, 2, 100)  # vzorová data

plt.figure(figsize=(5, 2.7), layout='constrained')

# vykreslení dat na implicitní osy
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

# popisky
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()                # legenda
```

```python
x = np.linspace(0, 2, 100)  # vzorová data

# přístup k obrazu a osám
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
# (Figure = kontejner nejvyšší úrovně pro všechny prvky grafu)

# vykreslení dat na osy
ax.plot(x, x, label='linear')
ax.plot(x, x**2, label='quadratic')
ax.plot(x, x**3, label='cubic')

# popisky
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title("Simple Plot")
ax.legend()                 # legenda
```

```python
# Fixing random state for reproducibility
np.random.seed(19680801)

# make up some data in the open interval (0, 1)
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))

# plot with various axes scales
plt.figure()

# linear
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)

# log
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)

# symmetric log
plt.subplot(223)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthresh=0.01)
plt.title('symlog')
plt.grid(True)

# logit
plt.subplot(224)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)
# Adjust the subplot layout, because the logit one may take more space
# than usual, due to y-tick labels like "1 - 10^{-3}"
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
                    wspace=0.35)

plt.show()
```

## Polární grafy

```python
theta = np.linspace(0.01, 2*np.pi, 150)     # úhel v polárním grafu
radius = np.log(theta)                      # vzdálenost od středu

ax = plt.subplot(projection="polar")
ax.plot(theta, radius)                      # vykreslení průběhu funkce v polárním grafu
 
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, subplot_kw=dict(projection='polar'))
ax1.plot(x, y)
ax2.plot(x, y ** 2)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0.01, 2*np.pi, 150)     # úhel v polárním grafu

radius1 = theta                             # první funkce: vzdálenost od středu
radius2 = 2*np.abs(theta-np.pi)             # druhá funkce: vzdálenost od středu
radius3 = 2*np.log(theta)                   # třetí funkce: vzdálenost od středu
 
ax = plt.subplot(111, projection="polar")

ax.plot(theta, radius1, "r.", label="f1")   # vykreslit průběh první funkce v polárním grafu
ax.plot(theta, radius2, "g", label="f2")    # vykreslit průběh druhé funkce v polárním grafu
ax.plot(theta, radius3, "b--", label="f3")  # vykreslit průběh třetí funkce v polárním grafu

plt.legend(loc="lower left")                # přidání legendy
 
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
# úhel v polárním grafu
theta = np.linspace(0.01, 4*np.pi, 150)
 
# první funkce: vzdálenost od středu
radius1 = theta
 
# druhá funkce: vzdálenost od středu
radius2 = 3*np.abs(theta-2*np.pi)
 
ax = plt.subplot(111, projection="polar")
 
# vykreslit průběh první funkce
# v polárním grafu
ax.plot(theta, radius2, "b", label="f1")
 
# vykreslit průběh druhé funkce
# v polárním grafu
ax.fill(theta, radius1, "yellow", alpha=0.3, label="f1")
 
# přidání legendy
plt.legend(loc="lower left")
 
# zobrazení grafu
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import font_manager as fm
 
 
# make a square figure and axes
fig = plt.figure(1, figsize=(6, 6), dpi=50)
ax = fig.add_axes([0.16, 0.16, 0.68, 0.68])
 
plt.title("Scripting languages")
ax.title.set_fontsize(30)
 
# popisky jednotlivých výřezů
labels = ['Perl', 'Python', 'Ruby']
 
# šířky jednotlivých výřezů
fracs = [90, 150, 70]
 
# vytáhnutí výřezů
explode = (0.0, 0.0, 0.15)
 
# barvy
colors = ('yellow', '#60ff60', 'red')
 
# vytvoření koláčového grafu
patches, texts, autotexts = ax.pie(fracs, explode=explode, colors=colors,
                                   labels=labels, autopct='%1.1f%%',
                                   shadow=True)
 
# změna stylu písma
proptease = fm.FontProperties()
proptease.set_size('xx-large')
plt.setp(autotexts, fontproperties=proptease)
plt.setp(texts, fontproperties=proptease)
 
plt.show()
```

Nastavení odchylek (chyb)

```python
# první pole hodnot a pole odchylek
vals1 = [10, 15, 20, 12, 14, 8]
delta1 = [1, 2, 3, 4, 5, 0]
 
# druhé pole hodnot a pole odchylek
vals2 = [19, 18,  6, 11,  6, 14]
delta2 = [4, 2, 3, 2, 2, 4]
 

N = len(vals1)                          # počet prvků
indexes = np.arange(N)                  # indexy prvků
width = 0.30                            # šířka sloupců
 
# sloupcový graf se dvěma skupinami sloupců
plt.bar(indexes, vals1, width, color='gray', edgecolor='black', label='CPU#1',
        yerr=delta1)
 
# posunuté sloupce
plt.bar(indexes+width, vals2, width, color='red', edgecolor='black',
        label='CPU#2', yerr=delta2)

plt.grid(True)                          # povolení zobrazení mřížky
plt.legend(loc="lower right")           # přidání legendy
 
plt.show()
```
