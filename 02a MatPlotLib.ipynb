Zobrazení os, mřížky, nastavení rozsahu, popisky

```python
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, "b-", label="sin")
plt.plot(x, y2, "r-", label="cos")

plt.legend(loc="lower left")                    # přidání legendy

plt.axis([-1, 8, -1.5, 1.5])                    # nastavení rozsahů (xmin, xmax, ymin, ymax)

plt.grid(True)                                  # povolení zobrazení mřížky
plt.axhline(y = 0, color='k')                   # vykreslení osy x (černě)
plt.axvline(x = 0, color='k')                   # vykreslení osy y (černě)

plt.xlabel("x")                                 # popis osy x
plt.ylabel("sin(x) a cos(x)")                   # popis osy y

# vložit první popisek do grafu
plt.annotate("maximální hodnota sin(x)",
             xy=(np.pi/2, 1.0),                 # poloha šipky (x, y)
             xytext=(1, 1.3),                   # poloha textu (x, y)
             arrowprops=dict(arrowstyle="->"))  # typ šipek
 
# vložit druhý popisek do grafu
plt.annotate("minimální hodnota cos(x)",
             xy=(np.pi, -1.0),                  # poloha šipky (x, y)
             xytext=(2, -1.3),                  # poloha textu (x, y)
             arrowprops=dict(arrowstyle="<-"))  # typ šipek
 
plt.show()
```

#### Použití implicitních objektů
globální proměnné (jako MATLAB)

```python
x = np.linspace(0, 2, 100)  # vzorová data

plt.figure(figsize=(6, 4.2), layout='constrained')  
# figsize je velikost obrázku v palcích; zde jako parametr figure
# layout='constrained' minimalizuje překryvy mezi objekty popisky

# vykreslení dat na implicitní osy
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

# popisky
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()                # legenda
plt.show()
```

Příklad s objektovým přístupem

```python
x = np.linspace(0, 2, 100)  # vzorová data

# přístup k obrazu a osám
fig, ax = plt.subplots(figsize=(6, 4.2), layout='constrained')  # velikost obrázku v palcích figsize jako parametr subplots

# vykreslení dat na osy
ax.plot(x, x, label='linear')
ax.plot(x, x**2, label='quadratic')
ax.plot(x, x**3, label='cubic')

# popisky
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title("Simple Plot")
ax.legend()                 # legenda
plt.show()
```

#### Nastavení různých měřítek
- linear
- log    - nelze reprezentovat nulové nebo záporné hodnoty (v 0 -oo)
- symlog - kladné i záporné hodnoty
- logit  - od 0 do 1

```python
# Nastavujeme náhodný generátor tak, aby byl výstup reprodukovatelný
np.random.seed(19680801)

# Generujeme 1000 náhodných hodnot z normálního rozdělení s průměrem 0.5 a odchylkou 0.4
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]        # Filtrujeme hodnoty, aby byly mezi 0 a 1
y.sort()                        # Seřazení hodnot y vzestupně
x = np.arange(len(y))           # Vytváříme vektor x s hodnotami od 0 do délky y-1

# plot with various axes scales
plt.figure()                    # Vytvoření nové figure pro vykreslení grafů

# První graf: Lineární měřítko pro osu y
plt.subplot(221)                # 2x2 mřížka, první graf
plt.plot(x, y)
plt.yscale('linear')            # Nastavujeme lineární měřítko pro osu y
plt.title('linear')             # Nadpis grafu
plt.grid(True)

# Druhý graf: Logaritmické měřítko pro osu y
plt.subplot(222)                # 2x2 mřížka, druhý graf
plt.plot(x, y)
plt.yscale('log')               # Nastavujeme logaritmické měřítko pro osu y
plt.title('log')                # Nadpis grafu
plt.grid(True)

# Třetí graf: Symetrické logaritmické měřítko pro osu y
plt.subplot(223)                # 2x2 mřížka, třetí graf
plt.plot(x, y - y.mean())       # Odečteme průměr hodnot pro posunutí do středu
plt.yscale('symlog', linthresh=0.01)    # Nastavujeme symetrické logaritmické měřítko
plt.title('symlog')             # Nadpis grafu
plt.grid(True)

# Čtvrtý graf: Logit měřítko pro osu y
plt.subplot(224)                # 2x2 mřížka, čtvrtý graf
plt.plot(x, y)
plt.yscale('logit')             # Nastavujeme logit měřítko pro osu y
plt.title('logit')              # Nadpis grafu
plt.grid(True)
# Adjust the subplot layout, because the logit one may take more space
# than usual, due to y-tick labels like "1 - 10^{-3}"
# Úprava pozic grafů na ploše (okraje a mezery mezi grafy)
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25, wspace=0.35)

plt.show()
```

#### Histogram

```python
from matplotlib import pyplot as plt
x = np.linspace(0, 1, 1000)
y = x ** 1.5
fig, axs = plt.subplots(1, 2)
fig.set_size_inches(10,5)           # velikost
axs[0].plot(x,y)                    # graf funkce
axs[1].hist(y)                      # histogram
plt.show()
```

#### Použití pro 3D grafy
Příklad: Vlevo graf funkce $z = \sin(\sqrt{x^2 + y^2})$, doprava testovací data

```python
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np

from mpl_toolkits.mplot3d.axes3d import get_test_data

fig = plt.figure(figsize=plt.figaspect(0.5))        # obr. 2x širší než vyšší

# =============
# 1. subplot
# =============
ax = fig.add_subplot(1, 2, 1, projection='3d')      # osy

X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)                            # 3D povrch
print("X: ", X)
print("Y: ", Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
ax.set_zlim(-1.01, 1.01)
fig.colorbar(surf, shrink=0.5, aspect=10)

# ==============
# 2. subplot
# ==============
ax = fig.add_subplot(1, 2, 2, projection='3d')      # osy

X, Y, Z = get_test_data(0.05)   
ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10)  # 3D drátěný model

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definice parametrů
u = np.linspace(0, 2 * np.pi, 50)  # u od 0 do 2π
v = np.linspace(-np.pi, np.pi, 50)  # v od -π do π
U, V = np.meshgrid(u, v)  # Vytvoření 2D gridu

# Parametrické rovnice
X = np.cos(U)
Y = np.sin(U) + np.cos(V)
Z = np.sin(V)

# Vykreslení plochy
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

ax.plot_wireframe(X, Y, Z, color='b', linewidth=0.5)  # Mřížkový model (mesh)
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Parametrická plocha")

plt.show()
```

```python
# kód je možné zkopírovat do prostředí pythonu
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

fig = plt.figure(figsize = (6,4))               # velikost obrázku
ax = plt.axes([0.1, 0.15, 0.8, 0.7])            # prostor grafu (rel. zleva, zdola, šířka, výška)
ax.set_ylim(-1.2, 1.2)                          # meze na ose y

# slider:
sld_ax = plt.axes([0.2, 0.9, 0.6, 0.05])        # prostor slideru (začíná o trochu víc vpravo, je nahoře, 60% šířky, 5% výšky)
sld = Slider(sld_ax, r'$\omega$ [Hz]', 1., 5., valinit=1.)
omega = sld.val                                 # přečtení hodnoty ze slideru

x = np.linspace(-2*np.pi, 2*np.pi, 800)
lines, = ax.plot(x, np.sin(2.*np.pi*omega*x))   # vykreslení počátečního tvaru křivky

def update_frequency(omega):                    # funkce při změně hodnoty slideru
    lines.set_ydata(np.sin(2.*np.pi*omega*x))   # aktualizace dat na ose y (nová data jako argument)

sld.on_changed(update_frequency)
plt.show()
```

Stejný příklad, funkce s parametry

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

# funkce s parametry
def f(t, A, f):
    return A * np.sin(2 * np.pi * f * t)

t = np.linspace(0, 1, 1000)
A = 5           # amplituda
init_f = 3      # počáteční frekvence

# graf a křivka
fig, ax = plt.subplots()
ax.set_xlabel('Time [s]')
line, = ax.plot(t, f(t, A, init_f), lw=2)   # n-tice s jedním prvkem

# posunutí grafu (kvůli místu pro posuvníky)
fig.subplots_adjust(left=0.1, bottom=0.25)

# posuvník pro frekvenci
axfreq = fig.add_axes([0.2, 0.1, 0.65, 0.03]) # od x, y, šířka, výška
f_slider = Slider(
    ax=axfreq,          # osa (obdélník, kam se bude vykreslovat)
    label='f [Hz]',     # popis
    valmin=0.1,         # minimální hodnota
    valmax=30,          # maximální hodnota
    valinit=init_f,     # počáteční hodnota
)

# funkce volaná při změně hodnoty slideru
def update(val):
    line.set_ydata(f(t, A, f_slider.val))
    fig.canvas.draw_idle()

f_slider.on_changed(update)  # registrace funkce update

plt.show()
```

Více posuvníků

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider, Button

def f(t, amplitude, frequency):     # parametrizovaná funkce
    return amplitude * np.sin(2 * np.pi * frequency * t)

t = np.linspace(0, 1, 1000)         # čas

init_amplitude = 5                  # počáteční hodnoty
init_frequency = 3

fig, ax = plt.subplots()            # vytvoření obrazu a křivky k manipulaci
line, = ax.plot(t, f(t, init_amplitude, init_frequency), lw=2)
ax.set_xlabel('Time [s]')

# uvolnění velikosti hlavního grafu pro slidery
fig.subplots_adjust(left=0.25, bottom=0.25)

# horizontální slider pro frekvenci
axfreq = fig.add_axes([0.25, 0.1, 0.65, 0.03])
freq_slider = Slider(
    ax=axfreq,
    label='Frequency [Hz]',
    valmin=0.1,
    valmax=30,
    valinit=init_frequency,
)

# vertikální slider pro amplitudu
axamp = fig.add_axes([0.1, 0.25, 0.0225, 0.63])
amp_slider = Slider(
    ax=axamp,
    label="Amplitude",
    valmin=0,
    valmax=10,
    valinit=init_amplitude,
    orientation="vertical"
)

def update(val):                        # při změně sliderů
    line.set_ydata(f(t, amp_slider.val, freq_slider.val))
    fig.canvas.draw_idle()

freq_slider.on_changed(update)          # registrace funkcí pro oba slidery
amp_slider.on_changed(update)

# tlačítko `matplotlib.widgets.Button` pro resetování sliderů do defaultních poloh
resetax = fig.add_axes([0.8, 0.025, 0.1, 0.04])
button = Button(resetax, 'Reset', hovercolor='0.975')


def reset(event):
    freq_slider.reset()
    amp_slider.reset()
button.on_clicked(reset)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact

# Definice funkce, která vykreslí graf podle zadaného parametru
def plot_graph(a):
    x = np.linspace(0, 2*np.pi, 100)
    y = np.sin(a*x)
    
    plt.plot(x, y)
    plt.title(f'Graf funkce sin({a}x)')
    plt.xlabel('x')
    plt.ylabel('sin(ax)')
    plt.grid(True)
    plt.show()

# Vytvoření slideru a jeho propojení s funkcí pro vykreslení grafu
interact(plot_graph, a=(1, 5, 0.1))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
#from mpl_toolkits.mplot3d import Axes3D
import ipywidgets as widgets

# Funkce pro vykreslení 3D parametrické plochy
def plot_surface(resolution=50):
    u = np.linspace(0, 2 * np.pi, resolution)
    v = np.linspace(-np.pi, np.pi, resolution)
    U, V = np.meshgrid(u, v)

    X = np.cos(U)
    Y = np.sin(U) + np.cos(V)
    Z = np.sin(V)

    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_wireframe(X, Y, Z, color='b', linewidth=0.5)

    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("Z")
    ax.set_title("Interaktivní 3D plocha")
    plt.show()

# Ovládání rozlišení pomocí slideru
slider = widgets.IntSlider(min=10, max=100, step=5, value=50, description="Rozlišení")
widgets.interactive(plot_surface, resolution=slider)
```
