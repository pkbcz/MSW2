```python
# kód je možné zkopírovat do prostředí pythonu
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

fig = plt.figure(figsize = (4,2))
ax = plt.axes([0.1, 0.15, 0.8, 0.7]) # axes for the plot
x = np.linspace(-2*np.pi, 2*np.pi, 800)
ax.set_ylim(-1.2, 1.2)

# Now, in the next step, we add a second axes object and put it into it a slider:
sld_ax = plt.axes([0.2, 0.9, 0.6, 0.05]) # axes for slider
sld = Slider(sld_ax, '$\omega$ [Hz]', 1., 5., valinit=1.)
omega=sld.val

# Plot of the initial curve
lines, = ax.plot(x, np.sin(2.*np.pi*omega*x))

def update_frequency(omega):
    lines.set_ydata(np.sin(2.*np.pi*omega*x))
sld.on_changed(update_frequency)
plt.show()
```

```python
# kód je možné zkopírovat do prostředí pythonu
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

fig = plt.figure(figsize=(4, 4))
sld_ax = fig.add_axes([0.2, 0.9, 0.6, 0.05])  # axes for slider
ax = fig.add_axes([0.1, 0.15, 0.8, 0.7])  # axes for the plot
ax.set_xlabel('Time [s]')
ax.set_ylabel('Amplitude [m]')

sld = Slider(sld_ax, '$\omega$ [Hz]', 1., 5., valinit=1.5)
omega = sld.val
x = np.linspace(-2*np.pi, 2*np.pi, 800)
ax.set_ylim(-1.2, 1.2)

# Plot of the initial curve
lines, = ax.plot(x, np.sin(2.*np.pi*omega*x), label=f'$\sin(2\pi\; {omega} x)$ ')
ax.legend()

def update_frequency(val):
    omega = sld.val
    lines.set_ydata(np.sin(2.*np.pi*omega*x))
    ax.legend((lines, ), (f'$\sin(2\pi\; {omega} x)$',))

sld.on_changed(update_frequency)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact

# Definice funkce, která vykreslí graf podle zadaného parametru
def plot_graph(a):
    x = np.linspace(0, 2*np.pi, 100)
    y = np.sin(a*x)
    
    plt.plot(x, y)
    plt.title(f'Graf funkce sin({a}x)')
    plt.xlabel('x')
    plt.ylabel('sin(ax)')
    plt.grid(True)
    plt.show()

# Vytvoření slideru a jeho propojení s funkcí pro vykreslení grafu
interact(plot_graph, a=(1, 5, 0.1))
```

```python
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, "b-", label="sin")
plt.plot(x, y2, "r-", label="cos")

plt.legend(loc="lower left")                    # přidání legendy

plt.axis([-1, 8, -1.5, 1.5])                    # nastavení rozsahů (xmin, xmax, ymin, ymax)

plt.grid(True)                                  # povolení zobrazení mřížky
plt.axhline(y = 0, color='k')                   # vykreslení osy x (černě)
plt.axvline(x = 0, color='k')                   # vykreslení osy y (černě)

plt.xlabel("x")                                 # popis osy x
plt.ylabel("sin(x) a cos(x)")                   # popis osy y

# vložit první popisek do grafu
plt.annotate("maximální hodnota sin(x)",
             xy=(np.pi/2, 1.0),                 # poloha šipky (x, y)
             xytext=(1, 1.3),                   # poloha textu (x, y)
             arrowprops=dict(arrowstyle="->"))  # typ šipek
 
# vložit druhý popisek do grafu
plt.annotate("minimální hodnota cos(x)",
             xy=(np.pi, -1.0),                  # poloha šipky (x, y)
             xytext=(2, -1.3),                  # poloha textu (x, y)
             arrowprops=dict(arrowstyle="<-"))  # typ šipek
 
plt.show()
```

```python
# Fixing random state for reproducibility
np.random.seed(19680801)

# make up some data in the open interval (0, 1)
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))

# plot with various axes scales
plt.figure()

# linear
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)

# log
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)

# symmetric log
plt.subplot(223)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthresh=0.01)
plt.title('symlog')
plt.grid(True)

# logit
plt.subplot(224)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)
# Adjust the subplot layout, because the logit one may take more space
# than usual, due to y-tick labels like "1 - 10^{-3}"
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
                    wspace=0.35)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0.01, 2*np.pi, 150)     # úhel v polárním grafu

radius1 = theta                             # první funkce: vzdálenost od středu
radius2 = 2*np.abs(theta-np.pi)             # druhá funkce: vzdálenost od středu
radius3 = 2*np.log(theta)                   # třetí funkce: vzdálenost od středu
 
ax = plt.subplot(111, projection="polar")

ax.plot(theta, radius1, "r.", label="f1")   # vykreslit průběh první funkce v polárním grafu
ax.plot(theta, radius2, "g", label="f2")    # vykreslit průběh druhé funkce v polárním grafu
ax.plot(theta, radius3, "b--", label="f3")  # vykreslit průběh třetí funkce v polárním grafu

plt.legend(loc="lower left")                # přidání legendy
 
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
# úhel v polárním grafu
theta = np.linspace(0.01, 4*np.pi, 150)
 
# první funkce: vzdálenost od středu
radius1 = theta
 
# druhá funkce: vzdálenost od středu
radius2 = 3*np.abs(theta-2*np.pi)
 
ax = plt.subplot(111, projection="polar")
 
# vykreslit průběh první funkce
# v polárním grafu
ax.plot(theta, radius2, "b", label="f1")
 
# vykreslit průběh druhé funkce
# v polárním grafu
ax.fill(theta, radius1, "yellow", alpha=0.3, label="f1")
 
# přidání legendy
plt.legend(loc="lower left")
 
# zobrazení grafu
plt.show()
```

Poznámka: meshgrid (mřížka sítě)
- parametr dvojice vektorů x (velikosti m) a y (velikosti n)
- vrátí dvě matice o n řádcích a m sloupcích
- v první tvoří řádky n vektorů x
- v druhé tvoří sloupce m vektorů y

```python
s = np.linspace(-1, 1, num = 3)         # -1, 0, 1

x, y = np.meshgrid(s, s)                # vrátí 2 matice
print("x:\n", x)                        # v každém řádku jsou čísla od -1 do 1
print("y:\n", y)                        # v každém sloupci jsou čísla od -1 do 1
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

# funkce s parametry
def f(t, A, f):
    return A * np.sin(2 * np.pi * f * t)

t = np.linspace(0, 1, 1000)
A = 5           # amplituda
init_f = 3      # počáteční frekvence

# graf a křivka
fig, ax = plt.subplots()
ax.set_xlabel('Time [s]')
line, = ax.plot(t, f(t, A, init_f), lw=2)   # n-tice s jedním prvkem

# posunutí grafu (kvůli místu pro posuvníky)
fig.subplots_adjust(left=0.1, bottom=0.25)

# posuvník pro frekvenci
axfreq = fig.add_axes([0.2, 0.1, 0.65, 0.03]) # od x, y, šířka, výška
f_slider = Slider(
    ax=axfreq,          # osa (obdélník, kam se bude vykreslovat)
    label='f [Hz]',     # popis
    valmin=0.1,         # minimální hodnota
    valmax=30,          # maximální hodnota
    valinit=init_f,     # počáteční hodnota
)

# funkce volaná při změně hodnoty slideru
def update(val):
    line.set_ydata(f(t, A, f_slider.val))
    fig.canvas.draw_idle()

f_slider.on_changed(update)  # registrace funkce update

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider, Button


# The parametrized function to be plotted
def f(t, amplitude, frequency):
    return amplitude * np.sin(2 * np.pi * frequency * t)

t = np.linspace(0, 1, 1000)

# Define initial parameters
init_amplitude = 5
init_frequency = 3

# Create the figure and the line that we will manipulate
fig, ax = plt.subplots()
line, = ax.plot(t, f(t, init_amplitude, init_frequency), lw=2)
ax.set_xlabel('Time [s]')

# adjust the main plot to make room for the sliders
fig.subplots_adjust(left=0.25, bottom=0.25)

# Make a horizontal slider to control the frequency.
axfreq = fig.add_axes([0.25, 0.1, 0.65, 0.03])
freq_slider = Slider(
    ax=axfreq,
    label='Frequency [Hz]',
    valmin=0.1,
    valmax=30,
    valinit=init_frequency,
)

# Make a vertically oriented slider to control the amplitude
axamp = fig.add_axes([0.1, 0.25, 0.0225, 0.63])
amp_slider = Slider(
    ax=axamp,
    label="Amplitude",
    valmin=0,
    valmax=10,
    valinit=init_amplitude,
    orientation="vertical"
)


# The function to be called anytime a slider's value changes
def update(val):
    line.set_ydata(f(t, amp_slider.val, freq_slider.val))
    fig.canvas.draw_idle()


# register the update function with each slider
freq_slider.on_changed(update)
amp_slider.on_changed(update)

# Create a `matplotlib.widgets.Button` to reset the sliders to initial values.
resetax = fig.add_axes([0.8, 0.025, 0.1, 0.04])
button = Button(resetax, 'Reset', hovercolor='0.975')


def reset(event):
    freq_slider.reset()
    amp_slider.reset()
button.on_clicked(reset)

plt.show()
```
